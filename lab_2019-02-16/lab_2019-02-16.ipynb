```python
 from sklearn.tree import DecisionTreeClassifier
```

```python
cols_to_analyze = {'landmass':['bars','stripes','colours','red','green','blue','gold','white','black','orange','mainhue','circles','crosses','saltires','quarters','sunstars','crescent','triangle','icon','animate','text','topleft','botright'], \
                   'language':['bars','stripes','colours','red','green','blue','gold','white','black','orange','mainhue','circles','crosses','saltires','quarters','sunstars','crescent','triangle','icon','animate','text','topleft','botright'], \
                   'religion': ['bars','stripes','colours','red','green','blue','gold','white','black','orange','mainhue','circles','crosses','saltires','quarters','sunstars','crescent','triangle','icon','animate','text','topleft','botright']}
flagdata_analyze = flagdata[[]]
for col in cols_to_analyze:
    columns = cols_to_analyze[col]
    clf=DecisionTreeClassifier( max_depth = 3 )
    clf.fit(flagdata[columns], flagdata[col])
    #print (col, clf.feature_importances_)
    feat_importance = dict( zip(columns, clf.feature_importances_) )
    print ( col, "---", sorted(feat_importance.items(), key=lambda x: x[1]) )
    print ('--')
   
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
k_range = list(range(1, 10))

cols_to_analyze = {'landmass':['bars','stripes','sunstars'], \
                   'language':['bars','stripes','sunstars'], \
                   'religion':['bars','stripes','sunstars']}

for col in cols_to_analyze:
    columns = cols_to_analyze[col]
    k_scores = []
    for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k)
        scores = cross_val_score(knn, flagdata[columns], flagdata[col], cv=3, scoring='accuracy')
        k_scores.append(scores.mean())
    print(col,k_scores)
```

```python
import operator
```

```python
cols_to_analyze = {'landmass':['bars','stripes','colours','red','green','blue','gold','white','black','orange','mainhue','circles','crosses','saltires','quarters','sunstars','crescent','triangle','icon','animate','text','topleft','botright'], \
                   'language':['bars','stripes','colours','red','green','blue','gold','white','black','orange','mainhue','circles','crosses','saltires','quarters','sunstars','crescent','triangle','icon','animate','text','topleft','botright'], \
                   'religion': ['bars','stripes','colours','red','green','blue','gold','white','black','orange','mainhue','circles','crosses','saltires','quarters','sunstars','crescent','triangle','icon','animate','text','topleft','botright']}

for maincol in cols_to_analyze:
    print (f'###\n###{maincol}\n###')
    for maincol_value in sorted( flagdata[maincol].unique() ):
        print (f'---{maincol} = {maincol_value}')
        tmpflagdata = flagdata.loc[flagdata[maincol] == maincol_value]
        tmpflagdata = tmpflagdata[cols_to_analyze[maincol]]
        tmpflagdata.reset_index()
        for col in tmpflagdata.columns:
            piv={}
            for i in range(0,len(tmpflagdata)):
                #print (len(tmpflagdata), col, i)
                for val in tmpflagdata[col]:
                    #print (val)
                    if str(val) in piv:
                        piv[str(val)]+=1
                    else:
                        piv[str(val)] = 1
            #piv_sorted = sorted(piv, key=lambda x: x[1])
            piv_sorted = sorted(piv.items(), key=operator.itemgetter(1))
            print (f'!!!---!!! {col} - {piv_sorted}')
```
